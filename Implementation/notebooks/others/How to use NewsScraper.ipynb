```python
import NewsScraper
import Preprocessor
import FileManager
```

# NewsScraper module
- module scrapes covid 19 articles 

## get_google_page_urls() function
- gets the URL of a google advanced search



```python
google_page_urls = NewsScraper.get_google_page_urls(NewsScraper.DOMAINS,NewsScraper.KEYWORDS)
```

```python
google_page_urls
```

## get_article_urls_from_google_page()
- scrapes the urls of the search results on the google page



```python
article_urls = []
for p_ in google_page_urls:
    article_urls.extend(NewsScraper.get_article_urls_from_google_page(NewsScraper.DOMAINS,p_))
```

```python
import random
sample_article_urls = random.sample(article_urls,20)
sample_article_urls
```

## scrape()
- scrapes all the text from the url 



```python
articles_list = []
for url in sample_article_urls:
    articles_list.append(NewsScraper.scrape(url,Preprocessor.cleanText))
```

```python
# save some articles for QA annotation
folder = r"C:\Users\tanch\Documents\GitHub\Covid-19-QA-System\QA System\Raw Data\articles"
for i,text in enumerate(articles_list):
    FileManager.saveTextFile(folder,i,text)
```
