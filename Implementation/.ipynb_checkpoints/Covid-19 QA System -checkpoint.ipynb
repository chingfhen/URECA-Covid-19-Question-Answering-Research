```python
import sys
sys.path.append(r"C:\Users\tanch\Documents\GitHub\Covid-19-QA-System\QA System\Custom Modules")
import Config
from haystack.document_store.elasticsearch import ElasticsearchDocumentStore
from haystack.retriever.sparse import ElasticsearchRetriever
from haystack.reader.farm import FARMReader
from haystack.pipeline import ExtractiveQAPipeline
from haystack.utils import print_answers
import pandas as pd
```

```python
# base_model= 'deepset/roberta-base-squad2'
# tuned_model_path = r"C:\Users\tanch\Documents\URECA\projects\Covid QA\Models\Trained models"+"\\roberta"
```

```python
reader = FARMReader(model_name_or_path=Config.MODEL_NAME,
                        context_window_size = 500,
                        max_seq_len = 100,
                        doc_stride =  30)
#tuned_model = FARMReader(model_name_or_path=tuned_model_path)
```

```python
document_store = ElasticsearchDocumentStore(index = Config.INDEX_NAME,
                                            username = Config.AUTH['username'],
                                            password = Config.AUTH['password'],
                                            host = "localhost",
                                            port = 9200,
                                            similarity = "dot_product",
                                            search_fields = ["text",'name'],
                                            text_field = "text",
                                            name_field = "name",
                                            embedding_field = "embedding",
                                            embedding_dim = 768)
retriever = ElasticsearchRetriever(document_store=document_store)
pipe = ExtractiveQAPipeline(reader, retriever) 
```

```python
questions = ["capacity for playing badminton","singing class max students","Max people in a tour","What are the covid-19 vaccines used by Singapore?","How many people can enter a tennis court?","Why did Singapore choose Pfizer-BioNTech vaccine?","Can i go overseas?","What are the punishments for breaking covid-19 rules twice?","Who can can get vaccinated in singapore?","Who were the first people to be vaccinated in singapore?","Why are healthcare professionals vaccinated first?","What is the minimum age to be vaccinated?","When did singapore get the Pfizer-BioNTech vaccine?","how much money was funded to the Jobs and Skills Package?","What is the Jobs and Skills Package for?","how many people are allowed at marriage solemnisations?","How many people can play in a basketball court?","Max people at a church","marriage solemnisations disallowed"]
answers = []
probabilities = []
contexts = []
```

```python
for question in questions:
    output = pipe.run(query=question, top_k_retriever=10, top_k_reader=1)
    answer,probability, context = output['answers'][0]['answer'],output['answers'][0]['probability'],output['answers'][0]['context']
    answers.append(answer)
    probabilities.append(probability)
    contexts.append(context)
```

```python
pd.DataFrame({'questions':questions,
              'answers':answers,
              "probabilities":probabilities,
              "contexts":contexts})
```
